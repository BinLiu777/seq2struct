```python
import json
import os
import re
import glob

import pandas as pd
```

```python
results_0214 = {}
for d in sorted(glob.glob('../logdirs/20190214/*')):
    exp_name = os.path.basename(d)
    eval_paths = glob.glob(os.path.join(d, 'eval-val-step*-bs1.jsonl'))
    results_0214[exp_name] = results_for_exp = {}
    for eval_path in eval_paths:
        step = int(re.search('step(\d+)', eval_path).group(1))
        scores = json.load(open(eval_path))
        results_for_exp[step] = scores
    print(d)
```
