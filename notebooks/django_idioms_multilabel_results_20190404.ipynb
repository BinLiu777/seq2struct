```python
import itertools
import json
import os

import natsort
import pandas as pd

from seq2struct.utils import evaluation
```

```python
os.chdir('..')
```

```python
base = 'logdirs/20190404-django-allmatches-anysplit-multimean/'
results = []
for section, dirname, bs in itertools.product(
         ('val',), natsort.natsorted(os.listdir(base)), (1,)):
    try:
        metrics = evaluation.compute_metrics(
            'configs/django/nl2code.jsonnet', '', section, 
            os.path.join(base, dirname, 'infer-{}-step2600.jsonl'.format(section)))
    except IndexError:
        metrics = None
    if metrics is None:
        continue
    
    results.append((dirname.split('/')[-1], section, bs) + tuple(metrics.values()))
```

```python
metrics.keys()
```

```python
df = pd.DataFrame(results, columns=('Model', 'Section', 'Beam size', 'Exact match',))
```
