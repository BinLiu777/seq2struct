```python
import json
import os
import re
import glob

import pandas as pd
```

```python
rows = []
for d in sorted(glob.glob('../logdirs/20190302/*')):
    exp_name = os.path.basename(d)
    output_from, upd_steps, upd_type, upd_tied, max_steps, batch_size = re.match(
        'output_from=([^,]+),upd_steps=(\d+),upd_type=([^,]+),upd_tied=([^,]+),max_steps=(\d+),batch_size=(\d+)', exp_name).groups()

    eval_paths = glob.glob(os.path.join(d, 'eval-val-step*-bs1.jsonl'))
    all_scores = []
    best = (0, 0)
    for eval_path in eval_paths:
        step = int(re.search('step(\d+)', eval_path).group(1))
        scores = json.load(open(eval_path))
        #best = max(best, (scores['total_scores']['all']['exact'], step))
        all_scores.append((scores['total_scores']['all']['exact'], step))
    all_scores.sort(reverse=True)
    if len(all_scores) == 0:
        all_scores = [(0, 0)]
    if len(all_scores) < 2:
        all_scores += all_scores
    rows.append((output_from, upd_steps, upd_type, upd_tied, max_steps, batch_size, len(eval_paths),
                 all_scores[0][1], all_scores[0][0], all_scores[1][1], all_scores[1][0]))
    print(d)
df = pd.DataFrame(rows, columns=('output_from', 'qenc', 'ctenc', 'upd_steps', 'max_steps', 'batch_size', 'num steps eval', 'step', 'acc', 'step', 'acc'))
```
