```python
import json
import os
import re
import glob

import pandas as pd
```

```python
results_0214 = {}
for d in sorted(glob.glob('../logdirs/20190214/*')):
    exp_name = os.path.basename(d)
    eval_paths = glob.glob(os.path.join(d, 'eval-val-step*-bs1.jsonl'))
    results_0214[exp_name] = results_for_exp = {}
    for eval_path in eval_paths:
        step = int(re.search('step(\d+)', eval_path).group(1))
        scores = json.load(open(eval_path))
        results_for_exp[step] = scores
    print(d)
```

```python
for exp in (
    'output_from=false,qenc=eb,ctenc=ebs,tinc=true,upd_steps=0',
    'output_from=false,qenc=eb,ctenc=ebs,tinc=false,upd_steps=1',
    'output_from=true,qenc=eb,ctenc=ebs,tinc=true,upd_steps=0',
    'output_from=true,qenc=eb,ctenc=ebs,tinc=false,upd_steps=1',
):
    rows = []
    rows.extend((exp, k, v) for k, v in results_0214[exp].items())
    
    best = max(rows, key=lambda x: x[2]['total_scores']['all']['exact'])
    scores = [best[2]['total_scores'][sec]['exact'] for sec in ('easy', 'medium', 'hard', 'extra', 'all')]
    print(best[0], best[1], ' & '.join('{:.2f}\%'.format(s * 100) for s in scores))
```

```python
results = {}
for d in sorted(glob.glob('../logdirs/20190223/*')):
    exp_name = os.path.basename(d)
    eval_paths = glob.glob(os.path.join(d, 'eval-val-step*-bs1.jsonl'))
    results[exp_name] = results_for_exp = {}
    for eval_path in eval_paths:
        step = int(re.search('step(\d+)', eval_path).group(1))
        scores = json.load(open(eval_path))
        results_for_exp[step] = scores
    print(d)
```

```python
for exp in (
    'output_from=false,qenc=eb,ctenc=ebs,upd_steps=2',
    'output_from=false,qenc=eb,ctenc=ebs,upd_steps=3',
    'output_from=false,qenc=eb,ctenc=ebs,upd_steps=4',
    'output_from=false,qenc=eb,ctenc=ebs,upd_steps=5',
    'output_from=true,qenc=eb,ctenc=ebs,upd_steps=2',
    'output_from=true,qenc=eb,ctenc=ebs,upd_steps=3',
    'output_from=true,qenc=eb,ctenc=ebs,upd_steps=4',
    'output_from=true,qenc=eb,ctenc=ebs,upd_steps=5',
):
    rows = []
    for full_exp in ('{},{}'.format(exp, b) for b in (
        'max_steps=40000,batch_size=10',
        'max_steps=40000,batch_size=20',
        'max_steps=80000,batch_size=10',
        'max_steps=80000,batch_size=20',)):
        rows.extend((full_exp, k, v) for k, v in results[full_exp].items())
    
    best = max(rows, key=lambda x: x[2]['total_scores']['all']['exact'])
    scores = [best[2]['total_scores'][sec]['exact'] for sec in ('easy', 'medium', 'hard', 'extra', 'all')]
    print(best[0], best[1], ' & '.join('{:.2f}\%'.format(s * 100) for s in scores))
```

```python
for d in sorted(glob.glob('../logdirs/20190214/*')):
    exp_name = os.path.basename(d)
    
    best = (1e10, 0)
    for line in open(os.path.join(d, 'log.txt')):
        m = re.search('Step (\d+) stats, val: loss = ([\d\.]+)', line)
        if m:
            best = min(best, (float(m.group(2)), int(m.group(1))))
    
    print(exp_name, best)
```

```python
rows = []
for d in sorted(glob.glob('../logdirs/20190214/*')):
    exp_name = os.path.basename(d)
    output_from, q_enc, ct_enc, tinc, upd_steps = re.match(
        'output_from=([^,]+),qenc=([^,]+),ctenc=([^,]+),tinc=([^,]+),upd_steps=(\d+)', exp_name).groups()

    eval_paths = glob.glob(os.path.join(d, 'eval-val-step*-bs1.jsonl'))
    best = (0, 0)
    for eval_path in eval_paths:
        step = int(re.search('step(\d+)', eval_path).group(1))
        scores = json.load(open(eval_path))
        best = max(best, (scores['total_scores']['all']['exact'], step))
    rows.append((output_from, q_enc, ct_enc, tinc, upd_steps, len(eval_paths), best[1], best[0]))
    
    print(d)
```

```python
df = pd.DataFrame(rows, columns=('output_from', 'qenc', 'ctenc', 'tinc', 'upd_steps', 'num steps eval', 'step', 'acc'))
```

```python
rows = []
for d in sorted(glob.glob('../logdirs/20190223/*')):
    exp_name = os.path.basename(d)
    output_from, q_enc, ct_enc, upd_steps, max_steps, batch_size = re.match(
        'output_from=([^,]+),qenc=([^,]+),ctenc=([^,]+),upd_steps=(\d+),max_steps=(\d+),batch_size=(\d+)', exp_name).groups()

    eval_paths = glob.glob(os.path.join(d, 'eval-val-step*-bs1.jsonl'))
    all_scores = []
    best = (0, 0)
    for eval_path in eval_paths:
        step = int(re.search('step(\d+)', eval_path).group(1))
        scores = json.load(open(eval_path))
        #best = max(best, (scores['total_scores']['all']['exact'], step))
        all_scores.append((scores['total_scores']['all']['exact'], step))
    all_scores.sort(reverse=True)
    if len(all_scores) == 0:
        all_scores = [(0, 0)]
    if len(all_scores) < 2:
        all_scores += all_scores
    rows.append((output_from, q_enc, ct_enc, upd_steps, max_steps, batch_size, len(eval_paths),
                 all_scores[0][1], all_scores[0][0], all_scores[1][1], all_scores[1][0]))
    print(d)
df = pd.DataFrame(rows, columns=('output_from', 'qenc', 'ctenc', 'upd_steps', 'max_steps', 'batch_size', 'num steps eval', 'step', 'acc', 'step', 'acc'))
```
