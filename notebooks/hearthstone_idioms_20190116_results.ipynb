```python
import itertools
import json
import os

import natsort
import pandas as pd

from seq2struct.utils import evaluation
```

```python
base = 'logdirs/20190116/'
results = []
for section, dirname, bs in itertools.product(
         ('train', 'val',), natsort.natsorted(os.listdir(base)), (1, 15)):
    try:
        metrics = evaluation.compute_metrics(
            'configs/hearthstone/nl2code.jsonnet', section, 
            os.path.join(base, dirname, 'infer-{}-step2600-bs{}.jsonl'.format(section, bs)))
    except IndexError:
        metrics = None
    if metrics is None:
        continue
    
    results.append((dirname[27:], section, bs) + tuple(metrics.values()))
```

```python
for section in ('train', 'val'):
    metrics = evaluation.compute_metrics(
        'configs/hearthstone/nl2code.jsonnet', section, 
        'logdirs/20190101/nl2code-hearthstone-dropout-fef2c5b/infer-{}-step2600-bs15-modeleval.jsonl'.format(section))
    results.append(('nl2code-hearthstone-dropout-20190101', section, 15) + tuple(metrics.values()))
```

```python
for section, bs in itertools.product(('train', 'val'), (1, 15)):
    metrics = evaluation.compute_metrics(
        'configs/hearthstone/nl2code.jsonnet', section, 
        'logdirs/20181228/nl2code-hearthstone-dropout-43628aa/infer-{}-step2600-bs{}-modeleval.jsonl'.format(section, bs))
    results.append(('nl2code-hearthstone-dropout-20191228', section, bs) + tuple(metrics.values()))
```

```python
df = pd.DataFrame(results, columns=('Model', 'Section', 'Batch size', 'Exact match', 'Sentence BLEU', 'Corpus BLEU'))
```
